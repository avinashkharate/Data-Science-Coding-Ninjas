```python
import imp
import matplotlib.pyplot as plt
import numpy as np
```

```python
sizes = [3,4,6,2]
plt.pie(sizes)
plt.show()
```

```python
sizes = [3,4,6,2]
colors = ['blue','purple','red','pink']
plt.pie(sizes,colors=colors)
plt.show()
```

```python
sizes = [3,4,6,2]
labels = ['A','B','C','D']
colors = ['blue','purple','red','pink']

plt.pie(sizes,colors=colors, labels=labels,autopct='%0.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
sizes = [3,4,6,2]
labels = ['A','B','C','D']
colors = ['blue','purple','red','pink']
explode = [0.1,0,0,0]

plt.pie(sizes,colors=colors, labels=labels,autopct='%0.1f%%',explode=explode)
plt.title('Pie Chart')
plt.show()
```

```python
sizes = [3,4,6,2]
labels = ['A','B','C','D']
colors = ['blue','purple','red','pink']
explode = [0.1,0,0,0]

plt.pie(sizes,colors=colors, labels=labels,autopct='%0.1f%%',explode=explode,counterclock=False)
plt.title('Pie Chart')
plt.show()
```

```python
sizes = [3,4,6,2]
labels = ['A','B','C','D']
colors = ['blue','purple','red','pink']
explode = [0.1,0,0,0]

plt.pie(sizes,colors=colors, labels=labels,autopct='%0.1f%%',explode=explode,counterclock=False,startangle=100)
plt.title('Pie Chart')
plt.show()
```
