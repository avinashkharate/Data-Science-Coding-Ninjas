```python

import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,5,0.1)
y2=x**2
y=x**3

# plt.plot(x,y,color='black',marker='o',)
plt.plot(x,y,color='red',linewidth=2,label='X^3')
plt.plot(x,y2,color='blue',linewidth=2,label="X^2")

# plt.axis([0,10,0,200])

plt.grid()

plt.text(2,80,'text')

plt.ylabel('y')
plt.xlabel("x")
plt.legend()
plt.title("Demo")
plt.show()
```
