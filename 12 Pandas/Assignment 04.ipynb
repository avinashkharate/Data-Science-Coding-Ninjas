Problem-4 : Terror Government
Problem Statement : There was formation of new government in India on 26 May 2014. So current government's span is from 26th May 2014 to current. Find out two things from this period-

1.Total number of attacks done in this period in India. Find this count as integer.
2. Which Terrorist group was most active in this period in India. Most active means, group which has done maximum number of attacks.
3. Ignore the Unknown group.

```python
import pandas as pd
df = pd.read_csv("terrorismData.csv")
df.head()
```

```python
df = df[df['Year']>=2014 ]
df =df[df['Month']>=5 ]
df =df[df['Day']>=26]
df = df[df['Country']=='India']
df.head()
```

```python
ans1 =df.shape[0]
print(ans1)
```

```python
df_group = df['Group'].value_counts
df_group
```

```python
df = df[df.Group!='Unknown']
temp = df['Group'].value_counts().index
ans2=temp[0]
print(ans2)
```

```python
print(ans1,ans2)
```
